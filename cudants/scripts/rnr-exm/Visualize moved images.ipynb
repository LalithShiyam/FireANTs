```python
import h5py
%pylab
import seaborn as sns
import os.path as osp
%matplotlib inline
```

```python
ROOT_DIR = "/data/rohitrango/RnR-ExM/test"

def get_images(pair):    
    moved_image_f = osp.join(ROOT_DIR, f"zebrafish_pair{pair}_warp.h5")
    moved_image_f = h5py.File(moved_image_f, "r")
    moved_image = moved_image_f['warped'][:]
    moved_image_f.close()
    # load fixed and moving images
    images_f = h5py.File(osp.join(ROOT_DIR, f"zebrafish_pair{pair}.h5"), 'r')
    fixed = images_f['fixed'][:]
    moving = images_f['move'][:]
    images_f.close()
    return fixed, moving, moved_image
```

```python
fixed, moving, warped = get_images(5)
```
